poission distribution

```python
from scipy.stats import poisson
```

```python
poisson.cdf(12,15) #exactly 12 incidents taking place today
```

```python
poisson.cdf(12,15)  #upto 12 incidents take place
```

```python
1-poisson.cdf(11,15)# minimum 12 incidents take place
```

```python
poisson.sf(11,15)  #is equal to 1-poisson.cdf(11,15)
```
