```python
#2-proportion test
#This test is used to compare the difference of two proportions calculated from two samples against a claim value and establish whether their difference is statistically significant or not
```

```python
engg = np.array([40,48])

total = np.array([70,80])
#clame value 8%
cm = 0.08

statsmodels.stats.proportion.proportions_ztest(engg,total,cm,alternative='smaller')
```

```python
from scipy.stats import chisquare
```

```python
counts = np.array([50,20,45,25])

pattern = np.array([0.4,0.1,0.3,0.2])

exp_counts = sum(counts) * pattern
```

```python
scipy.stats.chisquare(counts,f_exp = exp_counts)
```
