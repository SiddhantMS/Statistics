** Merging the data frame **\
we have pd.concat option but now we will do on pd.mearge()


```python
pune = pd.read_excel("CDAC_DataBook.xlsx",sheet_name='Pune')
mumbai = pd.read_excel("CDAC_DataBook.xlsx",sheet_name='Mumbai')
```

```python
pune
```

```python
mumbai
```

```python
#there is no row and column  matching
# there is no person which have same naem sub grade
pd.merge(pune,mumbai) #we will get empty df but there is no any matching of common coln or row entire row should be common name subject and grade should be same
```

```python
pd.merge(pune,mumbai,on='Subject')  #only common subject betn pune and mumbai will display
```

```python
pd.merge(pune,mumbai,on='Subject',how = 'inner')  #by default merger take as inner
```

```python
df1 = pd.merge(pune,mumbai,on='Subject',how = 'outer')
df1
```
